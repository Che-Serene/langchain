## 🎯 과제 1: 나만의 챗봇 만들기 (기본 체인 활용)

- **목표:** LangChain의 PromptTemplate, LLMChain 이해 및 응용
- ChatPromptTemplate을 사용해 사용자 질문을 요약하는 프롬프트 작성
- LLMChain을 구성해 “질문 요약 → 응답 생성” 체인 구현
- **주제: “학생 상담 챗봇”**

**유의 사항**
- 체인 구조의 올바른 구현 (PromptTemplate → LLMChain)
- 주제에 맞는 프롬프트 설계력

```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 모델 로드
llm = ChatOllama(model="exaone3.5:7.8b", temperature=0.7)

# 프롬프트 템플릿 설정 (선택 사항이지만 권장)
prompt = ChatPromptTemplate.from_messages([
    ("system", "사용자의 질문을 요약하여 출력하고 응답 생성"),
    ("user", "{input}")
])

# 출력 파서 설정 (LLM의 출력을 문자열로 파싱)
output_parser = StrOutputParser()

# 체인 구성
chain = prompt | llm | output_parser
```

```python
question = "대한민국의 수도는 어디고 인구는 몇 명인가요?"
response = chain.invoke({"input": question})
response
```

## 🧩 과제 2: 다중 체인 연결 (SequentialChain 실습)

- **목표:** 여러 체인을 순차적으로 연결하는 방법 학습
- 첫 번째 체인: 사용자 입력을 분석하여 요약문 생성
- 두 번째 체인: 요약문을 바탕으로 키워드 추출

- **두 체인을 SequentialChain으로 연결해 결과 출력**

**예시**

- 입력: “LangChain은 LLM 응용 프레임워크로, 프롬프트 관리와 체인 구성이 가능하다.”
- 출력: 요약 + 주요 키워드 (예: LangChain, LLM, 프롬프트, 체인)

**유의 사항**
- 체인 연결 구조 이해도
- 입력과 출력 간의 데이터 흐름 처리

```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 모델 로드
llm = ChatOllama(model="exaone3.5:7.8b", temperature=0.7)

# 프롬프트 템플릿 설정 (선택 사항이지만 권장)
prompt1 = ChatPromptTemplate.from_messages([
    ("system", "사용자의 질문을 요약합니다."),
    ("user", "{input}")
])

prompt2 = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant."),
    ("user", "{input}")
])

# 출력 파서 설정 (LLM의 출력을 문자열로 파싱)
output_parser = StrOutputParser()

# 체인 구성
chain1 = prompt | llm | output_parser
chain2 = chain1 | llm | output_parser
```

```python
question = "대한민국의 수도는 어디고 인구는 몇 명인가요?"
response = chain2.invoke({"input": question})

response
```

## 💬 과제 3: 커스텀 프롬프트 엔지니어링

- **목표:** 동일한 모델에 대해 프롬프트 설계가 결과에 미치는 영향 탐구


- 동일한 질문에 대해 3가지 다른 프롬프트 작성

- 예: “요약해줘”, “핵심 문장을 뽑아줘”, “학문적 요약으로 써줘”

- 각 결과를 비교하고, 어떤 프롬프트가 가장 일관성 있는지 확인

```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 모델 로드
llm = ChatOllama(model="exaone3.5:7.8b", temperature=0.7)

# 프롬프트 템플릿 설정 (선택 사항이지만 권장)
prompt_a = ChatPromptTemplate.from_messages([
    ("system", "요약해줘."),
    ("user", "{input}")
])

prompt_b = ChatPromptTemplate.from_messages([
    ("system", "핵심 문장을 뽑아줘."),
    ("user", "{input}")
])

prompt_c = ChatPromptTemplate.from_messages([
    ("system", "학문적 요약으로 써줘."),
    ("user", "{input}")
])

# 출력 파서 설정 (LLM의 출력을 문자열로 파싱)
output_parser = StrOutputParser()

# 체인 구성
chain_a = prompt_a | llm | output_parser
chain_b = prompt_b | llm | output_parser
chain_c = prompt_c | llm | output_parser
```
